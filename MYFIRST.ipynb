list and nested list

```python
mylist = [1,2,'apple']
print(mylist[0]) #1
print(mylist[2]) #apple
mylist2 = [1,2,["hi","hello"],3]
print(mylist2[0]) #1
print(mylist2[2]) # hi,hello
print(mylist2[2][0]) #hi
mylist3 = [1,2,3,["hi","hello",["vanakkam","namaskar"],50],4]
print(mylist3[3][2][0]) #vanakkam
```

list : order, change, duplicate
append, remove, sort, copy-> since = only make a reference ,insert, pop, reverse

```python
a = [1,2,3] 
print(a[0]) #1

b = ["apple","banana", "apple"]
print(b) #prints with duplicates

c = ["hi", "hello"]
c.append("vanakkam")
print(c)     # ['hi', 'hello', 'vanakkam']

c.sort()
print(c)     # ['hello', 'hi', 'vanakkam']

c.remove("hi")
print(c)     # ['hello', 'vanakkam']

c.insert(0, "apple")
print(c)     # ['apple', 'hello', 'vanakkam']
```
